```python
# Cleaning each DataFrame (df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024)

# Create a mapping dictionary to standardize names
team_name_mapping = {
    "Brighton": "Brighton and Hove Albion",
    "Brighton & Hove Albion": "Brighton and Hove Albion",
    "Bournemouth": "AFC Bournemouth",
    "Man City": "Manchester City",
    "Man Utd": "Manchester United",
    "Newcastle Utd": "Newcastle United",
    "Nott'ham Forest": "Nottingham Forest",
    "Spurs": "Tottenham Hotspur",
    "Tottenham": "Tottenham Hotspur", 
    "West Brom": "West Bromwich Albion",
    "West Ham": "West Ham United",
    "Wolves": "Wolverhampton Wanderers"
    # Add more when finding discrepancies
}

df_all = [df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024]

# df_22_23_cleaned = clean_league_table(df_22_23, '2022-23') - didn't use this as it was duplicating the values of the 2020-2023 dataframe


def clean_league_table(df, season_name):
    """
    A function to clean each scraped league table DataFrame.
    """
    # 1. Clean column names
    df.columns = df.columns.str.lower()

    # 2. Drop fully NaN columns, only "notes" or "Notes"
    for col in ["notes", "Notes"]:
        if col in df.columns:
            df = df.drop(columns=[col])

    # 3. Clean Team names: Remove (C), (R), etc.
    if "team" in df.columns:
        df['team'] = df['team'].str.replace(r'\s*\([RC]\)', '', regex=True).str.strip()
        df['team'] = df['team'].replace(team_name_mapping)

    # 4. Rename possession column
    df = df.rename(columns={"possession": "poss", "possession (%)": "poss"})

    # 5. Add season column
    df['season'] = season_name

    # 6. Ensure numeric columns are truly numbers
    numeric_cols = ['gf', 'ga', 'poss']
    for col in numeric_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')

    # 7. GF/GA ratio if gf and ga exist
    if "gf" in df.columns and "ga" in df.columns:
        df["gf_ga_ratio"] = df["gf"] / df["ga"]

    return df
```

```python
# 2.A. Possession
# Q: Is possession correlated with league points?
# Null Hypothesis (H₀): There is no correlation between possession rate and league points/winning the league.
# Data: Kaggle API: average possession per team per match/season.
# Univariate analysis showing Possession distribution

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.histplot(df_concat["poss"], bins=20, kde=True)
plt.title("Distribution of Possession %")
plt.xlabel("Possession %")
plt.ylabel("Count")
plt.show()
```

```python
# Mapped results W/D/L/ to numerical points
df_concat_poss = df_concat.copy()  # or whichever DF has possession + results
df_concat_poss["points"] = df_concat_poss["result"].map({"W": 3, "D": 1, "L": 0})
```

```python
# Aggregated by season + team
# Total points per season and average possession:

df_possession_points = (
    df_concat_poss.groupby(["season", "team"])
    .agg(
        total_points=("points", "sum"),
        avg_poss=("poss", "mean")
    )
    .reset_index()
)
```

```python
# Average Possession per Team Calculated over 7 seasons 
# All Cleaned DFs have the column "poss", so concatenated in df_concat, mapped results for points, and agg by season + team in df_possession_points
df_possession_points 
```

```python
# Bivariate analysis showing a Scatterplot (Possession vs Points)

plt.figure(figsize=(8,6))
sns.scatterplot(data=df_possession_points, x="avg_poss", y="total_points")

# Add trendline
sns.regplot(data=df_possession_points, x="avg_poss", y="total_points", scatter=False, color="red")

plt.title("Possession % vs League Points")
plt.xlabel("Possession %")
plt.ylabel("League Points")
plt.show()
```

```python
from scipy.stats import pearsonr, spearmanr

# Dropped rows with NaN
df_corr = df_possession_points.dropna(subset=["avg_poss", "total_points"])

pearson_r, pearson_p = pearsonr(df_corr["avg_poss"], df_corr["total_points"])
spearman_rho, spearman_p = spearmanr(df_corr["avg_poss"], df_corr["total_points"])

print(f"Pearson r: {pearson_r:.20f}, p={pearson_p:.20f}")
print(f"Spearman rho: {spearman_rho:.20f}, p={spearman_p:.20f}")
```

```python
# Histogram: Shows whether possession % is normally distributed across teams.
# Scatterplot + Trendline: Quick visual: does higher possession usually mean more points?
# Pearson’s r: Measures linear correlation (values close to +1 = strong positive correlation).
# Spearman’s ρ: Checks if there’s a monotonic relationship (increasing trend, even if not perfectly linear).
# p-values: If < 0.05, the relationship is statistically significant.
```

```python
# 2.B. Possession for Home vs Away Histograms
# Question: Is possession percentage higher at home or away games?
# Null Hypothesis (H₀): There is no correlation between possession rate and playing games at home.
# Data: Kaggle API: average possession per team per match/season.

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

# Histogram
sns.histplot(data=df_concat, x="poss", hue="venue", bins=20, alpha=0.3, element="step")

plt.title("Possession % Distribution: Home vs Away")
plt.xlabel("Possession %")
plt.ylabel("Density / Count")
plt.show()
```

```python
# Histogram
# The orange bars are away games, and the blue bars are home games.
# Both distributions are centered roughly around 50% possession.
# There’s a lot of overlap between home and away.
# However, the blue (home) distribution is slightly shifted to the right → meaning teams tend to have a little more possession at home.
```

```python
# 2.B. Possession for Home vs Away KDEs 
sns.kdeplot(data=df_concat, x="poss", hue="venue", fill=True, common_norm=False, alpha=0.4)
```

```python
from scipy.stats import ttest_ind

# Home and away possession split into separate series
home_poss = df_concat[df_concat["venue"] == "Home"]["poss"].dropna()
away_poss = df_concat[df_concat["venue"] == "Away"]["poss"].dropna()

# Independent t-test run
t_stat, p_val = ttest_ind(home_poss, away_poss)

print("T-statistic:", t_stat)
print("P-value:", p_val)

# Result
if p_val < 0.05:
    print("Reject H₀: There IS a significant difference in possession between home and away.")
else:
    print("Fail to reject H₀: No significant difference found.")
```

```python
# 1. Histogram

# The orange bars are away games, and the blue bars are home games.
# Both distributions are centered roughly around 50% possession.
# There’s a lot of overlap between home and away.
# However the blue (home) distribution is slightly shifted to the right, meaning teams tend to have a little more possession at home.

# 2. KDE (smoothed distribution curve)

# Both curves peak around the same place (~50–55% possession).
# The home curve (blue) is just a bit higher on the right side (55–65%) compared to away.
# The away curve (orange) is a bit higher on the left side (~40–50%).


# Null Hypothesis (H₀): There is no correlation between possession rate and playing at home.
# From your plots, possession is slightly higher at home, but not dramatically.

# Conclusion:
# The data suggests a small home advantage in possession.
# But since the overlap is large, the difference may not be statistically significant, will test with a t-test to be sure.
# Teams usually have about the same possession at home and away, but on average, they tend to keep the ball a little more when playing at home.
```
