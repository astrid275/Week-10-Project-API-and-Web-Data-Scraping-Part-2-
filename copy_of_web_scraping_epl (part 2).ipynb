```python
# Cleaned each DataFrame (df_18, df_19, df_20, df_21, df_22, df_23, df_24, df_25)

# Created a mapping dictionary to standardize names
team_name_mapping = {
    "Brighton": "Brighton and Hove Albion",
    "Brighton & Hove Albion": "Brighton and Hove Albion",
    "Bournemouth": "AFC Bournemouth",
    "Man City": "Manchester City",
    "Man Utd": "Manchester United",
    "Newcastle Utd": "Newcastle United",
    "Nott'ham Forest": "Nottingham Forest",
    "Spurs": "Tottenham Hotspur",
    "Tottenham": "Tottenham Hotspur", 
    "West Brom": "West Bromwich Albion",
    "West Ham": "West Ham United",
    "Wolves": "Wolverhampton Wanderers"
    # Added more as I found more discrepancies
}

def clean_league_table(df, season_name):
    """
    A function to clean each scraped league table DataFrame.
    """
    # 1. Cleaned Team names: Removed (C), (R), etc.
    df['team'] = df['team'].str.replace(r'\s*\([RC]\)', '', regex=True).str.strip()
    
    # 2. Added a 'Season' column to identify the year
    df['season'] = season_name # e.g., '2019-20'
    
    # 3. Standardized team names using team_name_mapping
    df['team'] = df['team'].replace(team_name_mapping)
    
    # 4. Converted to numeric (in case they were strings)
    numeric_cols = ['pld', 'w', 'd', 'l', 'gf', 'ga', 'gd', 'pts']
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce') # 'coerce' turns bad text into NaN

    # 5. Added GF/GA ratio as a column to answer the Business Question
    df["gf_ga_ratio"] = df["gf"] / df["ga"]

    # 6. Added "top3" or "non-top3" as a column to answer the Business Question
    if "top3" not in df.columns:
        df["top3"] = df_all["pos"].astype(int).apply(lambda x: "Top 3" if x <= 3 else "Non-Top 3")
    
    # 7. Calculated GD, ensuring negative values become 0 and not NaN
    df['gd'] = (df['gf'] - df['ga']).clip(lower=0)

    
    return df

# Cleaned dfs for all seasons below

df_18_cleaned = clean_league_table(df_18, '2017-18')
df_19_cleaned = clean_league_table(df_19, '2018-19')
df_20_cleaned = clean_league_table(df_20, '2019-20')
df_21_cleaned = clean_league_table(df_21, '2020-21')
df_22_cleaned = clean_league_table(df_22, '2021-22')
df_23_cleaned = clean_league_table(df_23, '2022-23')
df_24_cleaned = clean_league_table(df_24, '2023-24')
df_25_cleaned = clean_league_table(df_25, '2024-25')
# df_26_cleaned = clean_league_table(df_26, '2025-26')

all_dfs = [df_18_cleaned, df_19_cleaned, df_20_cleaned, df_21_cleaned, df_22_cleaned, df_23_cleaned, df_24_cleaned, df_25_cleaned]
```

```python
# Pearson Correlation test: GF/GA vs Points

from scipy.stats import pearsonr
import numpy as np

# Drop rows where either gf_ga_ratio or pts is NaN/Inf
df_valid = df_all[["gf_ga_ratio", "pts"]].replace([np.inf, -np.inf], np.nan).dropna()

gfga = df_valid["gf_ga_ratio"]
points = df_valid["pts"]

if gfga.nunique() > 1 and points.nunique() > 1:
    r, p_val = pearsonr(gfga, points)
    print("Pearson correlation r:", r)
    print("P-value:", p_val)
else:
    print("Pearson correlation is undefined (constant values or no valid data).")

# Interpretation:
# r close to 1 = strong positive correlation → higher GF/GA predicts more points.
# p_val < 0.05 = statistically significant.
    
```

```python
# T-test / Mann–Whitney Top 3 vs Non Top 3Rest

from scipy.stats import ttest_ind, mannwhitneyu

top3 = df_all[df_all["top3"] == "Top 3"]["gf_ga_ratio"].dropna()
non_top3 = df_all[df_all["top3"] == "Non-Top 3"]["gf_ga_ratio"].dropna()

# T-test
t_stat, p_val_t = ttest_ind(top3, non_top3)
print("T-test p-value:", p_val_t)

# Mann-Whitney (non-parametric backup)
u_stat, p_val_u = mannwhitneyu(top3, non_top3)
print("Mann-Whitney p-value:", p_val_u)

# Interpretation:
# If either test returns p < 0.05, then GF/GA ratio is significantly higher for Top 3 teams compared to others.
```
